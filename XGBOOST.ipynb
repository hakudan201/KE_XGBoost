```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./loan_data.csv")
```

```python
df
```

```python
df.dropna(inplace=True) #remove rows with missing values
df.drop(columns=['Loan_ID'], inplace=True) #delete column Loan_ID
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
is_Category = df.dtypes == object #checks which columns have a data type of 'object' (categorical)
category_column_list = df.columns[is_Category].tolist() #creates a list of those column names
df[category_column_list] = df[category_column_list].apply(lambda col: le.fit_transform(col)) #transforms categorical values into numerical labels
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2, random_state=42) #split train and test data size 4:1, random state value make each split the same
x_train = train.drop(columns=['Loan_Status'])
y_train = train['Loan_Status']

x_test = test.drop(columns=['Loan_Status'])
y_test = test['Loan_Status']
```

```python
from sklearn.ensemble import GradientBoostingClassifier
# Gradient Boosting
model = GradientBoostingClassifier(learning_rate=0.01, random_state=42, n_estimators = 100) #data set is not large so we use 0.01
model.fit(x_train, y_train)
```

```python
acc = model.score(x_test, y_test)
print(acc*100)
```

```python
import xgboost as xgb
import matplotlib.pyplot as plt
# XGBoost
model_xgb = xgb.XGBClassifier(random_state=42, n_estimators = 100, learning_rate=0.01) #set the number of boosting rounds (trees) in the ensemble
# more tree increase training time and performance
model_xgb.fit(x_train, y_train)
```

```python
acc = model_xgb.score(x_test, y_test)
print(acc*100)
```
